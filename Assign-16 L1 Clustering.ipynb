```python
from sklearn.cluster import KMeans
```

```python
wcss=[]

for i in range(1,11):
    km=KMeans(n_clusters=i)
    km.fit_predict(data1)
    wcss.append(km.inertia_)
```

```python
plt.plot(range(1,11),wcss)
plt.title('Elbow Method')
plt.Xlable('Number Of Clusters')
plt.Ylable('WCSS')
```
